# Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler, \
                                PowerTransformer, \
                                MinMaxScaler, \
                                QuantileTransformer 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

pd.set_option('display.max_columns', None)
```

```python
weather_data_df = pd.read_csv("weather_data.csv")
bike_data_df = pd.read_csv("bike_trip_data.csv")
need_to_pred_df = pd.read_csv("predictions.csv",header=None)
```

```python
need_to_pred_df
```

# Weather DataFrame

```python
weather_data_df
```

```python
columns_for_use = []
for column in weather_data_df.columns:
    if "Avg" in column:
        columns_for_use.append(column)
    if "Precipitation_Inches" in column:
        columns_for_use.append(column)
    if "Events" in column:
        columns_for_use.append(column)
    if "Date" in column:
        columns_for_use.append(column)
```

```python
columns_for_use
```

```python
weather_data_df["Date"] = weather_data_df["Date"].astype('datetime64[D]').astype(int)

weather_data_df = weather_data_df[columns_for_use]

weather_data_df = weather_data_df.replace("-",0)

weather_data_df["Precipitation_Inches"] = weather_data_df["Precipitation_Inches"].replace("T", 0)

weather_data_df["Events"] = weather_data_df["Events"].replace(np.nan,"No Event")

for column in weather_data_df.columns:
    if "Date" in column or "Events" in column:
        continue
    weather_data_df[column] = weather_data_df[column].astype(str).astype(float)
    
weather_data_df = pd.get_dummies(weather_data_df)

weather_data_df
```

# Predictions DataFrame

```python
pred_df = need_to_pred_df.rename(columns={0:"Month", 1:"Day", 2:"Address", 3:"Prediction"})
addresses_to_predict = list(pred_df.Address.unique())
print(len(addresses_to_predict),addresses_to_predict)

pred_df = pred_df[~pred_df["Address"].str.contains("Convention")]
addresses_to_predict = list(pred_df.Address.unique())
print(len(addresses_to_predict),addresses_to_predict)
pred_df.insert(2, 'year', 2016)
pred_df.insert(0,"Date",pred_df[pred_df.columns[0:3]].apply(
    lambda x: '/'.join(x.dropna().astype(str)),
    axis=1
))
need_to_pred_df = pred_df.drop(columns=["Month","Day","year"])
pred_df = pred_df.drop(columns=["Month","Day","year"])
pred_df["Date"] = pred_df["Date"].astype('datetime64[D]').astype(int)
pred_df = pd.get_dummies(pred_df)
pred_df
```

# FINAL DF: Weather and Predictions

```python
pred_weather_df = pred_df.merge(weather_data_df, how='left', on="Date")
pred_weather_df = pred_weather_df.drop(columns="Prediction")
pred_weather_df
```

# Bike Data DataFrame

```python
bike_data_df["Target"] = 1 + bike_data_df["duration_minutes"] * .1

bike_data_df = bike_data_df[bike_data_df['start_station_name'].isin(addresses_to_predict)]

bike_data_df.insert(0,"Date",bike_data_df[bike_data_df.columns[0:3]].apply(
    lambda x: '/'.join(x.dropna().astype(str)),
    axis=1
))
bike_data_df = bike_data_df.drop(columns=["month","day","year"])

bike_data_df["Date"] = bike_data_df["Date"].astype('datetime64[D]').astype(int)

bike_data_df = bike_data_df[["Date","start_station_name","duration_minutes"]]

bike_data_df = pd.get_dummies(bike_data_df)
bike_data_df
```

# FINAL DF: Weather and Bike Data

```python
weather_bike_df = bike_data_df.merge(weather_data_df,how='left',on="Date")
weather_bike_df
```

```python
for i in range(weather_bike_df.shape[1]):
    sm.qqplot(weather_bike_df.iloc[:,i], line='45')
    plt.title(weather_bike_df.columns[i])
    plt.show()
    
```

# Machine Learning

```python
inputs = weather_bike_df.drop(columns="duration_minutes")
target = weather_bike_df["duration_minutes"]
feature_names = inputs.columns
inputs = inputs.values
```

```python
target.describe()
```

```python
fig, ax = plt.subplots(figsize = (6,4))
target.plot(kind = "box")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=.2,random_state=0)
```

```python
sc_y = StandardScaler()
sc_x = StandardScaler()

qt_x = QuantileTransformer(output_distribution="normal")
qt_y = QuantileTransformer(output_distribution="normal")

y_train_scaled = qt_y.fit_transform(y_train.values.reshape(-1,1))
y_test_scaled = qt_y.fit_transform(y_test.values.reshape(-1,1))

x_train[:,1:] = qt_x.fit_transform(x_train[:,1:])
x_test[:,1:] = qt_x.fit_transform(x_test[:,1:])
```

```python
x_train[0]
```

```python
sm.qqplot(y_train_scaled, line='45')
```

```python
for i in range(x_train.shape[1]):
    sm.qqplot(x_train[:,i], line='45')
    plt.title(feature_names[i])
    plt.show()
    
```

## Linear Regression

```python
lr = LinearRegression()
lr.fit(x_train, y_train_scaled)
lr.score(x_test, y_test_scaled)
```

## Lasso Regression

```python
lasso = Lasso(random_state = 0,max_iter = 10000)
lasso.fit(x_train, y_train_scaled)
lasso.score(x_test,y_test_scaled)
```

```python
for coef, feature in sorted(zip(lasso.coef_, feature_names),reverse=True):
    print(f'{round(coef,4)} :  {feature}')
```

## Random Forest Regressor

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [60, 80, 100],
    'max_features': range(2,6,1),
    'n_estimators': [100, 200, 300, 1000]
}
# Create a based model
rf = RandomForestRegressor(random_state=1)
# Instantiate the grid search model
grid = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 2, n_jobs = -1, verbose = 2)


# Fitting the grid search RFR model with the training data
grid_result = grid.fit(x_train, y_train_scaled)

# Printing the best score and parameter
print('Best Score: ', grid_result.best_score_)
print('Best Params: ', grid_result.best_params_)

```

```python
rfr = RandomForestRegressor(**grid_result.best_params_)
rfr.fit(x_train, y_train_scaled.reshape(-1,))
rfr.score(x_test, y_test_scaled.reshape(-1,))
```
